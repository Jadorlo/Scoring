```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Chargement des différents CSV en Dataframe

```python
df_data_OG = pd.read_csv('files/revenus.csv') # Data originales
df_data_V1 = pd.read_csv('files\data/data_V1.csv') # Data avec traitement des valeurs manquantes et aberrantes (avec les "?" et Never-Worked regroupées)
df_data_V2 = pd.read_csv('files/data/data_V2.csv') # Data avec traitement des valeurs manquantes et aberrantes + regroupements des modalités des variables qualis et classification des variables quantis
```

Definition des variables quali / quanti pour l'analyse descriptive

```python
# Il y a une séparation des listes des variables entre celle pour les Data Originales et les autres car partir des Data V1, les "age" et "hour-per-weeks" sont regroupées en classes d'ages.

# Listes pour le DATA_OG
variables_quali_DATA_OG = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country','hours-per-week','income']
variables_quali_sans_income_DATA_OG = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country',]
variables_quanti_DATA_OG = ['age','capital-gain', 'capital-loss', 'hours-per-week']

# Listes pour les autres DATA
variables_quali = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country','age','hours-per-week','income']
variables_quali_sans_income = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country','age','hours-per-week']
variables_quanti = ['capital-gain', 'capital-loss']
```

# 0. Statistiques Descriptives Data Originales

### 0.A. Graphiques Variables Qualitatifs

```python
# Boucle qui itere le processus de création d'un piechart pour chaque variables qualitatives
for col in variables_quali_DATA_OG:
    # Calcul les proportions d'appartitions des modalités dans la variable
    prop_mod = df_data_OG[col].value_counts()
    print(prop_mod)

    # Créer le pie chart
    plt.figure(figsize=(4, 4))
    plt.pie(prop_mod, labels=prop_mod.index, autopct='%1.1f%%', startangle=140)
    plt.title(f"Proportion des modalités de {col}")
    plt.show()
```

### 0.B. Graphiques Variables Quantitatifs

```python
# Boucle qui itere le processus de création d'une courbe de densité pour chaque variables quantitatives
for variable in variables_quanti_DATA_OG:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(df_data_OG[variable], shade=True)  # shade=True pour colorier sous la courbe
    plt.title(f'Courbe de densité de {variable}')
    plt.xlabel(variable)
    plt.ylabel('Densité')
    plt.show()
```

### 0.C. Boxplots Variables Quantitatives

```python
# Boucle qui itere le processus de création de boxplots pour chaque variables quantitatives
for variable in variables_quanti_DATA_OG:
    plt.figure(figsize=(10, 2))
    plt.title(f'Boxplot des valeurs de {variable}')
    sns.boxplot(x=df_data_OG[variable])
    plt.xlabel(variable)
    plt.show()
```

### 0.D. Statistiques Bivariées

```python
# Boucle qui itere le processus de création de diagrammes en barres empilées, de la proportion d'income pour chaque modalités de chaques variables qualitatives
for variable in variables_quali_sans_income_DATA_OG :
    #Calcule la proportion la proportion de >50k et de <=50k d'income
    grouped = df_data_OG.groupby([variable, 'income']).size().unstack()
    percentages = grouped.divide(grouped.sum(axis=1), axis=0) * 100
    print(percentages)

    # Tri de l'affichage des modalités dans l'ordre croissant en fonction du pourcentage de la classe '<=50K'
    percentages['sort'] = percentages['<=50K']
    percentages = percentages.sort_values(by='sort', ascending=False).drop(columns='sort')

    # Création du diagramme
    ax = percentages.plot(kind='barh', stacked=True, color=['#ec5757', '#4fe29d'])
    plt.title('Répartition des revenus (%) en fonction du type de '+ variable )
    plt.ylabel(variable)
    plt.xlabel('Pourcentage (%)')
    plt.legend(title='Income', loc='upper left')
    plt.show()
```
